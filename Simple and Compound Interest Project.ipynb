### While developing their business activity, a company or an individual may, at a certain point, need financial resources that they do not currently have available. Various financial institutions offer the opportunity to obtain the necessary capital in the form of a loan (credit). ## <p style="color: red"> **1. Simple interest**</p>

### Simple interest is the one in which in each period we only interest the initial capital $K$, but not the interest from the previous periods. In such a case, the interest for each of the periods during which the capital is used is the same:

#### $$L_1 = L_2 = ... = L_n = L_n = \frac{p}{100}K=\frac{pK}{100}$$

#### The interest for all $n$ periods is

#### $$L = L_1 + L_2 + ... + L_n=\frac{npK}{100}$$

#### and the accumulated capital for these $n$ periods:

#### $$S_n = K + L = K + \frac{n p K}{100} = K \bigg( 1 + \frac{n p}{100} \bigg)$$

#### We calculate the interest rate $L$ and the capital accumulated over $n$ years $S_n$ by simply replacing the interest rate $p$ for one period in the above formulas with $p_a$:

#### $$L = \frac{np_aK}{100},\quad  S_n = K\bigg( 1 + \frac{np_a}{100} \bigg)$$

#### Here are some examples with different monthly Initial capital of $K$ = 10000 BGN is invested for 5 years at $p_a$% = 7% annual interest rate. Let's calculate the accumulated capital $S_5$ after the 5-year period. We have

#### $$S_5 = 10000\bigg( 1 + \frac{5.7}{100} \bigg)= 10000\;.\; \frac{135}{100} = 13500$$

#### In case the interest is calculated for periods of one month, the interest rate $L$ and the accumulated capital $S_n$ for $n$ months are calculated by replacing the interest rate $p$ for one period in the above formulas with 
#### $$p_м = \frac{p_a}{12}$$

#### $$L = \frac{np_aK}{1200},\quad  S_n = K\bigg( 1 + \frac{np_a}{1200} \bigg)$$

```python
K = 10000      
p_a = 7        
months = 60    

S_yearly = [K * (1 + (n * p_a) / 1200) for n in range(months + 1)]
S_monthly = [K * (1 + (n * p_a) / 1200) for n in range(months + 1)]  


months_range = list(range(months + 1))

plt.plot(months_range, S_yearly, label='Simple interest (annual formula)', color='blue')
plt.plot(months_range, S_monthly, label='Simple interest (monthly compounding)', color='red', linestyle='dashed')

plt.title('Comparison: Simple Interest — Annual vs Monthly Formula')
plt.xlabel('Month')
plt.ylabel('Capital(BGN)')


plt.grid(True)
plt.xlim(0, months)
plt.ylim(K, max(S_yearly) + 500)

plt.xticks(range(0, months + 1, 6))

plt.legend()

plt.show()
```

## <p style="color: red"> **2. Nominal and real interest rate**</p>

- #### The **nominal** interest rate means the interest on 100 monetary units of credit for one compounding period. In fact, this is exactly the interest rate defined above. In financial mathematics, the nominal interest rate is usually used.

- #### The **real** interest rate is obtained from the nominal interest rate after taking into account the inflation rate and reflects the increase (respectively, decrease) in the purchasing power of money. We will now see that the real interest rate is not simply the difference between the nominal interest rate and the inflation rate.

#### The purchasing power of capital $K$ is the amount of goods that can be purchased at the beginning of the period with this capital.
#### For example:
#### If you have $K$ 5000 (BGN).
#### And one $C$ piece of product costs 50 (BGN).
#### How many $q$ pieces can you buy?
#### $$q = \frac{K}{C},\quad q =\frac{5000}{50} = 100$$

#### The most famous formula for the real interest rate is the **FISHER** formula:
#### $$r = 100\;.\;\frac{p-i}{100 + i}$$

- ### Example 1. **Inflation**
    #### Nominal interest rate: $𝑝 = 10$%
    #### Inflation: $𝑖 = 5$%
    #### $$r = 100\;.\;\frac{10-5}{100 + 5}\; = 100\;.\;\frac{5}{105}\; = 100\;.\;0,04762=4,76\%$$

- ### Example 2. **Deflation**
    #### Nominal interest rate: $𝑝 = 10$%
    #### Inflation: $𝑖 = -2$%
    #### $$r = 100\;.\;\frac{10-(-2)}{100 + (-2)}\;= 100\;.\;\frac{12}{98}\; = 100\;.\;0,1224=12,24\%$$

## <p style="color: red"> **3. Compound interest**</p>

#### In the case where the loan repayment period is divided into equal periods and the accumulated capital at the end of each period is used as the initial capital for the next period, we are talking about compound interest. In this way of interest calculation, the interest from each period is added to the capital, capitalized. Let capital $K$ be compounded at $p$% compound annual interest. The accumulated capital $S_1$ after one period will be:

#### $$S_1 = K\bigg( 1 + \frac{p}{100} \bigg)$$
#### We set $K_1$ = $S_1$ and after a few more periods we find that:
#### $$S_n = K\bigg( 1 + \frac{p}{100} \bigg)^n$$

### **Example** 1: Accumulation of 10000 (BGN) at 7% annual compound interest for 5 years
#### $$S_5 = 10000\;.\;\bigg( 1 + \frac{7}{100} \bigg)^5 = 10000\;.\;(1,07)^5$$
#### $$S_5 = 10000\;.\;1,40255=14025,5(BGN)$$

### **Example** 2: Accumulation of 5000 (BGN) at 5% annual compound interest for 3 years
#### $$S_3 = 5000\;.\;\bigg( 1 + \frac{5}{100} \bigg)^3 = 5000\;.\;(1,05)^3$$
#### $$S_3 = 5000\;.\;1,157625=5788,13(BGN)$$

### **Example** 3: If interest is capitalized every 6 months (half a year), at 8% annual interest rate, for 3 years
#### Here the number of periods is $𝑛 = 3 × 2 = 6$ (because we have 2 periods per year), and the interest rate for the period is $𝑝 = \frac{8}{2} = 4$%.
#### $$S_6 = K\bigg(1 + \frac{4}{100}\bigg)^6$$

#### If $𝐾$ = 2000, then:
#### $$S_6 = 2000\;.\;(1,04)^6=2000\;.\;1,265319=2530,64(BGN)$$

## <p style="color: red">**5. Annuity calculations**</p>
#### An annuity (from Latin annuitas - annual installment) is a financial transaction in which periodic equal payments (or installments) are made at regular intervals of time - for example, monthly, quarterly, annually.

### **Types of annuities according to the start time:**
#### - **Immediate annuity**
#### Payment of installments begins immediately after the conclusion of the contract.
#### - **Deferred annuity**
#### Payment of installments begins after a certain period from the conclusion of the contract.
### **Types of annuities according to the time of payment within each period:**
#### - **Prepaid annuity (Annuity Due)**
#### The payment of each installment is made at the beginning of the respective period.**
#### - **Post-period annuity (Ordinary Annuity)**
#### The payment of each installment is made at the end of the respective period.

### <p style="color: red">**Instalments**</p>
#### I will give some examples of installments with simple and compound interest.

### **Simple interest with periodic installments**
#### With simple interest, you pay an amount every year. Each installment bears interest for the remaining years until the end. The interest is not capitalized (i.e. you do not accumulate interest on interest).
#### $$S = n\;.\;a\;.\;\bigg(1+\frac{(n+1)p}{200}\bigg)$$

#### **Example**: If we deposit 10,000 (BGN) each year at the beginning. With an annual interest rate of 10% for the period of 5 years.
#### Where:
#### $n$ = 5
#### $a$ = 10000
#### $p$ = 10
#### $$\frac{(5+1)\;.\;10}{200}=\frac{60}{200}=0,3$$
#### $$1 + 0,3 = 1,3$$
#### $$5\;.\;10000\;.\;1,3=65000$$

### **Compound interest with periodic installments**
#### $$S = a\;.\;q\;.\;\frac{q^n-1}{q-1}$$
### **Example**:
#### Where:
#### $n$ = 5
#### $a$ = 10000
#### $q = 1 + \frac{p}{100}=1,1$
#### $$q^5 = 1,1^5\approx1,61051$$
#### $$q^5 - 1 = 0,61051$$
#### $$q-1 = 0,1$$
#### $$10000\;.\;1,1\;.\;\frac{0,61051}{0,1}=10000\;.\;1,1\;.\;6,1051=10000\;.\;6,7156=67156$$


```python
a = 10000
p = 10
n = 5
q = 1 + p/100

simple_sums = []
for year in range(1, n+1):
    total = 0
    for i in range(year):
        periods = n - i
        total += a * (1 + (periods * p) / 100)
    simple_sums.append(total)

compound_sums = []
for year in range(1, n+1):
    total = 0
    for i in range(year):
        periods = n - i
        total += a * (q ** periods)
    compound_sums.append(total)

years = list(range(1, n+1))

plt.plot(years, simple_sums, marker='o', color='green',label='Simple interest')
plt.plot(years, compound_sums, marker='s', color='blue',label='Compound interest')

for x, y in zip(years, simple_sums):
    plt.text(x+0.4, y-2000, f'{y:.2f}', fontsize=10, color='green', ha='center')

for x, y in zip(years, compound_sums):
    plt.text(x-0.4, y+100, f'{y:.2f}', fontsize=10, color='blue', ha='center')

plt.xlabel('Year')
plt.xticks(years)
plt.ylabel('Accumulated amount (BGN)')
plt.title('Accumulation with annual contributions')
plt.grid(True)

plt.legend(loc='upper left')
plt.show()
```

### <p style="color: red">**Repayments**</p>

#### Here we will consider issues related to the repayment of long-term loans. When drawing a   loan $K$, it is important to know what the annuity payment a will be for each period, how long (how many periods $n$) it will be repaid, and what amount $D$ in total will be returned to the lender at $p$% compound interest for one interest period. Let us set $q = 1 + \frac{p}{100}$ and calculate how the loan decreases over time and the annuities withheld. For the first period, the loan K will bear interest and at the end of the period the first payment will be made. Therefore, at the end of the first period there will be an amount of $D1 = Kq - a$ remaining to be paid.

### **Example**. A loan of 5000 (BGN) was taken out. with a repayment period of 4 years and an annual interest rate of 12%. Find the monthly annuity payment, the total amount due and the interest on the loan. Since the annual interest rate is 12% and the interest period is one month, the interest rate for one period (in this case, one month) is $p$% = 1%. Then $q = 1 +\frac{1}{100} = 1,01,\;n = 4.12 = 48$ the period and $K = 5000(BGN)$

#### $$a = \frac{5000\;.\;1,01^{48}\;.\;(1,01-1)}{1,01^{48}-1}=131,67(BGN)$$

#### For the repayment of the loan, the following will be extended:
#### $$S=48\;.\;131,67=6320,16(BGN)$$

#### and the interest rate on the loan is:
#### $$L = 6320,16 - 5000 = 1320,16(BGN)$$
